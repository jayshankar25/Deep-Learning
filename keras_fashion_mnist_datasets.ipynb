```python
# import modules

import tensorflow as tf
from tensorflow import keras
import  numpy as np
import  matplotlib.pyplot as plt
```

```python
# load / read the data from keras datasets 'fashion_mnist'

data = keras.datasets.fashion_mnist
data
```

```python
# split the data into training and testing in form of tuple


(train_images , train_labels) , (test_images , test_labels) = data.load_data()
```

```python
# show the datatype , size , length etc of the data

print(type(train_images))
print(len(train_images))
print(train_images.shape)
print(train_images.ndim)
```

```python
print(train_images)
print(len(train_images[0]))
print(train_images[0].shape)
```

```python
# data of image in form of matrix 28 * 28


print(train_images[0])   # matrix 28*28
```

```python
print(train_labels)
print(train_labels.shape)
print(train_labels.ndim)
```

```python
# convert the matrix into image 

plt.imshow(train_images[5] , cmap = plt.cm.binary)
plt.show
```

### Scaling the matrix between 0 and 1

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

```python
# after scaling the data between 0 and 1

print(train_images[0])
```

```python
plt.imshow(train_images[5], cmap = plt.cm.binary)
plt.show()
```

```python
class_names = ['T-Shirt/Top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankel Boot']
```

### Creating a model

```python

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28 , 28)) ,
    keras.layers.Dense(128 , activation = 'relu') , 
    keras.layers.Dense(10 , activation= 'softmax')
    
]) 


model.compile(optimizer = "adam" , loss = "sparse_categorical_crossentropy" , metrics = ["accuracy"])

model.fit(train_images , train_labels , epochs = 5 )

test_loss , test_accuracy = model.evaluate(test_images , test_labels)

print("Test Accuracy :- " , test_accuracy)

print("Test Loss :- " , test_loss)
```

### Using the model to make the prediction

```python
prediction = model.predict(test_images)
print(prediction)
```

```python
for i in range(5):
    plt.grid(False)
    plt.imshow(test_images[i] , cmap = plt.cm.binary)
    plt.xlabel("Actual : " + class_names[test_labels[i]] )
    plt.title("Predicted :- " + class_names[np.argmax(prediction[0])])
    plt.show()
```
